```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/cleaned_dataset.csv')
data.head(10)
```

```python
df = data.sort_values(by=['overall_experience'],  ascending=False)
```

```python
df['cleaned_resume'][0]
```

```python
#program to create get the Resumes based on category
def find_category(category):
    rslt_df = df.loc[df['Category'] == category]
    rslt_df = rslt_df.loc[:,["Name","cleaned_resume"]]
    category = rslt_df.iloc[0:20]
    return category
```
